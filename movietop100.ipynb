```python
import urllib, json

url = "https://raw.githubusercontent.com/hjorturlarsen/IMDB-top-100/master/data/movies.json"
response = urllib.request.urlopen(url)
data = json.loads(response.read())
print(data)
```

```python
with open('movie.json','w') as f:
    json.dump(data,f)
```
